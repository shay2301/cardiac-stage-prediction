```python
from IPython.display import HTML, display

def play_video_in_colab(video_path):
    """
    Embeds and plays a video file within a Colab notebook.

    Args:
        video_path (str): Path to the video file.
    """
    video_file = open(video_path, "rb").read()
    video_url = "data:video/mp4;base64," + b64encode(video_file).decode()

    display(HTML(f"""
        <video width=800 controls>
            <source src="{video_url}" type="video/mp4">
        </video>
    """))

# Example usage
```

```python
folder_path = '/content/drive/MyDrive/YData/Project/VIZ.AI project/EchoNet-Dynamic/Videos/'

filename = os.listdir(folder_path)[-1]
play_video_in_colab(folder_path+filename)
```
