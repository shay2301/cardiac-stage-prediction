```python
from google.colab import drive
drive.mount('/content/drive')
volumeTraces = pd.read_csv('/content/drive/MyDrive/YData/Project/VIZ.AI project/EchoNet-Dynamic/VolumeTracings.csv')
frame_labels = pd.read_csv('/content/drive/MyDrive/YData/Project/VIZ.AI project/EchoNet-Dynamic/Frames labels.csv')
```

```python

bound_box = volumeTraces.groupby(['FileName','Frame']).agg(  {'X1': ['min', 'max'],'Y1': ['min', 'max'],'X2': ['min', 'max'],'Y2': ['min', 'max']}).reset_index()

bound_box['max_X']  = bound_box.apply(lambda row: max(row['X1']['max'], row['X2']['max']), axis=1)
bound_box['max_Y']  = bound_box.apply(lambda row: max(row['Y1']['max'], row['Y2']['max']), axis=1)
bound_box['min_X']  = bound_box.apply(lambda row: min(row['X1']['min'], row['X2']['min']), axis=1)
bound_box['min_Y']  = bound_box.apply(lambda row: min(row['Y1']['min'], row['Y2']['min']), axis=1)
display(bound_box)
bound_box.drop(columns=['X1','X2','Y1','Y2'], inplace = True)
```

```python
filename = '0X4C017259B103AAEE.avi'
df = bound_box[bound_box['FileName'] ==  filename ]
display(df)
```
