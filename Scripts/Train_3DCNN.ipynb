```python
import  torch
from torchvision.models.video import r3d_18
from torchsummary import summary

# Load the R3D model
model = r3d_18(pretrained=True)

# Assuming the R3D model takes in a 3-channel video clip of dimensions 112x112 with a temporal depth of 16 frames
# and you are using a device like 'cuda' or 'cpu'
device = "cuda" if torch.cuda.is_available() else "cpu"
model.to(device)

# Use the torchinfo summary method to print a summary of the model
summary(model, input_size=( 3, 16, 112, 112), device=device)
```
